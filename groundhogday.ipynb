  ## So to begin, lets gather some data

```python
import numpy as np
import pandas as pd
import json
import csv
import matplotlib.pyplot as plt

# First lets get data for all predictive groundhogs. link: https://groundhog-day.com/api/v1/groundhogs/ 
json_filepath = 'ghogs.json'

with open(json_filepath, 'r', encoding="utf8") as json_file:
    # Load the JSON content
    data = json.load(json_file)

# heres our dataframe for all groundhogs and predictions (plus lots of other info)
groundhogs = pd.json_normalize(data['groundhogs'])
groundhogs.set_index(['id'], inplace=True) 

# Now lets get some weather data along with phils predictions. 
# link: https://www.kaggle.com/datasets/groundhogclub/groundhog-day?resource=download
csv_filepath = 'phil_and_weather.csv'
phil_weather = pd.read_csv(csv_filepath)
```

## Important Definitions
It is important to know exactly what an "Early Spring" actually means. Officially on the Punxsutawney Phil / Gobler's Knob website, they do not define what an early spring actually is. However, some meteorologists accept an early spring as being the period around the spring equinox having a temperature above 40 degrees Fahrenheit, which is what I will be going with in my analysis. The NOAA specifically states that it uses March average temperatures (https://www.noaa.gov/heritage/stories/grading-groundhogs). It should also be noted that when Phil sees his shadow on G-Day that means there will be 6 more weeks of winter. So if Phil see's no shadow, there will be an early spring.\
\
If we take a look at our phil and weather data:

```python
phil_weather
```

We can see that we have columns for Febraury and March average temperatures for different regions of the USA. Unfortunately, the first 9 years are missing data and therefore are not helpful in calculating Phil's predictive accuracy. For our general definition of early spring, we are going to need the average temperature from March for each year.

```python
# the last entry of phil weather is average across the 20th century so we are just going to drop it
phil_weather.drop(131, axis=0, inplace=True)

# get the mean spring temperature for each year
spring_avg = phil_weather['March Average Temperature']

# We can see that there were 86 years where there was an early spring nationwide (ignoring years with no data).
spring_avg[spring_avg > 40].count()
```

Now we have everything we need to figure out Pux Phil's national prognosticating average.

```python
# lets copy some data over to a new DF
simple_phil = phil_weather[['Year', 'Punxsutawney Phil']].copy()
simple_phil['March Avg Temp'] = spring_avg

# lets get rid of any years with no record or no temp
simple_phil.dropna(inplace=True)
simple_phil = simple_phil[~simple_phil['Punxsutawney Phil'].str.contains('No Record')]

# Which of phils predictions are correct?
simple_phil['Prediction Correct?'] = (simple_phil['Punxsutawney Phil'].str.contains('Full Shadow')) & (simple_phil['March Avg Temp'] < 40)

# What is his overall accuracy?
phil_accuracy = simple_phil['Prediction Correct?'].sum() / simple_phil['Prediction Correct?'].count() * 100
phil_accuracy
```

So we can see above that Punxsutawney Phil is not very good at predicting the weather, with an accuracy rating of 24%. This is slightly worse than the figure given by the NOAA which claims Phil has a 35% accuracy (Is this a government coverup to make Phil look better than he actually is?? Now the question is how does Phil compare to the rest of the prognosticating groundhogs? Looking at our groundhog data from groundhog-day.com, we have a lot of information, but we are mainly interested in only a few columns. This dataset also lists other weather predicting animals like squirrels.

```python
groundhogs.head()
```

So as you can see, this is a LOT of information, most of which we don't really need (If you are reading the PDF version of this, it will look terrible and take up a whole page, sorry). What we want to do here is get all those predictions into a neat dataframe along with some other useful information like the groundhog name and country.

```python
groundhogs_mini = groundhogs.reset_index()[['id', 'name', 'country', 'type']]

# the predictions are formatted like this:
# {year:
# shadow: {None, 0, 1}
# details:}
# we could organize this all via multiIndex like
#                    shadow
# hog     years       
records = []

for hog_id, yearly_predictions in groundhogs['predictions'].items():
    for pred in yearly_predictions:
        records.append({
            'hog_id': hog_id,
            'year': pred['year'],
            'shadow': pred['shadow']
        })

# Create DataFrame
hog_predictions = pd.DataFrame(records)

# merge so we can get names country and type
hogs = hog_predictions.merge(groundhogs_mini, left_on='hog_id', right_on='id')

# Set MultiIndex
hogs.drop('id', axis=1, inplace=True)
hogs.set_index(['hog_id', 'name', 'country', 'type'], inplace=True)

#sort the index
hogs.sort_index(inplace=True)
hogs
```

Wow look at that data! Now that this has all been combined into one convenient DataFrame, we can begin to do some more interesting comparisons and analyses. Lets add the weather data to our dataframe so that we can calculate each groundhogs accuracy.

```python
# lets add our weather data to the hogs dataframe

# First, prepare a DataFrame with just the year and March temperature
march_temps = phil_weather[['Year', 'March Average Temperature']].copy()
march_temps.rename(columns={'Year': 'year', 'March Average Temperature': 'march_avg_temp'}, inplace=True)

# Convert both 'year' columns to integers
hogs['year'] = hogs['year'].astype(int)
march_temps['year'] = march_temps['year'].astype(int)

# Merge with hogs DataFrame on 'year'
hogs = hogs.reset_index()  # Reset index to access 'year' column easily
hogs = pd.merge(hogs, march_temps, on='year', how='left')

# Restore MultiIndex
hogs.set_index(['hog_id', 'name', 'country', 'type'], inplace=True)
hogs.sort_index(inplace=True)
```

```python
# Our weather data only includes years from 1895 to 2016, so we are going to cut out everything that isn't within those years
# as well as any years where a groundhog's prediction wasn't recorded.
hogs.dropna(inplace=True)

# we can determine if each prediction was true or false
hogs['correct_prediction'] = (hogs['shadow'] == True) & (hogs['march_avg_temp'] < 40)

# now we can see each hogs accuracy
hogs.groupby(['name'])[['correct_prediction']].mean().sort_values(by='correct_prediction', ascending=False).head()
```

Finally we can put all this information into a nice bar chart to get a good understanding of our data! We will include the mean, standard deviation, and number of predictions for each hog in our graph.

```python
# Group by name and get mean, std, and count
hog_stats = hogs.groupby('name')['correct_prediction'].agg(['mean', 'std', 'count'])

# Filter out groundhogs with 0 mean accuracy
hog_stats = hog_stats[hog_stats['mean'] > 0]

# Sort by mean accuracy
hog_stats = hog_stats.sort_values('mean', ascending=False)

# Plot
plt.figure(figsize=(14, 7))
bars = plt.bar(
    hog_stats.index,
    hog_stats['mean'],
    yerr=hog_stats['std'],
    color='orange',
    capsize=3,
    error_kw=dict(ecolor='gray', lw=1, alpha=0.6)  # thin semi-transparent error bars
)

# Add number of predictions as labels above bars
for bar, count in zip(bars, hog_stats['count']):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/4, height + 0.02, f'{count}', ha='center', fontsize=14)

plt.title('Groundhog Prediction Accuracy (with Standard Deviation and Count)')
plt.ylabel('Accuracy')
plt.xlabel('Groundhog Name')
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 1.05)
plt.tight_layout()
plt.show()
```
